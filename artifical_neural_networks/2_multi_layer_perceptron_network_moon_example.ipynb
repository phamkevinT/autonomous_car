## Buiding a multi-layer perception network for binary classification - moon example

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets

noisy_moons = datasets.make_moons(n_samples=1500, noise=0.05)
```

```python
noisy_moons
```

```python
X, y = noisy_moons
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
# Make new class predictions for classification problem

import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import make_blobs
```

```python
# Divide data into training and test datasets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# MinMaxScaler used for feature scaling (0.0 to 1.0)

from sklearn.preprocessing import MinMaxScaler

# Scale training dataset
scalar = MinMaxScaler()
scalar.fit(X_train)
X_train_scaled = scalar.transform(X_train)

# Scale test dataset
scalar = MinMaxScaler()
scalar.fit(X_test)
X_test_scaled = scalar.transform(X_test)
```

```python
plt.scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=y_train)
```

```python
# Define and fit the final model

# 'dense' = fully connected network
# binary_crossentropy - output is binary

model = Sequential()
model.add(Dense(1, input_dim=2, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
model.fit(X_train_scaled, y_train, epochs=500, verbose=2)
```

### Trying additional neurons to improve results


```python
# Define and fit the final model

model = Sequential()

# Input Layer
model.add(Dense(20, input_dim=2, activation='relu'))

# Adding multiple hidden layers
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))

# Output layer
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
model.fit(X_train_scaled, y_train, epochs=500, verbose=2)
```

```python
model.summary()
```

```python
# Make prediction
y_predict = (model.predict(X_test_scaled) > 0.5).astype("int32")
```

```python
from sklearn.metrics import classification_report, confusion_matrix

cm = confusion_matrix(y_test, y_predict)

sns.heatmap(cm, annot=True, fmt="d")
```

```python
print(classification_report(y_test, y_predict))
```

```python
# Visualizing the training dataset results

from matplotlib.colors import ListedColormap

X_grid, y_grid = X_train_scaled, y_train
X1, X2 = np.meshgrid(np.arange(start = X_grid[:, 0].min() - 1, stop = X_grid[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_grid[:, 1].min() - 1, stop = X_grid[:, 1].max() + 1, step = 0.01))

plt.contourf(X1, X2, model.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_grid)):
    plt.scatter(X_grid[y_grid == j, 0], X_grid[y_grid == j, 1],
                color = ListedColormap(('red', 'green'))(i), label = j)
    
plt.title('Training dataset')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```
