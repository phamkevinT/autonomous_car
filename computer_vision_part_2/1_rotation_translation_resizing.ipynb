# LOAD/DISPLAY IMAGE

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline

image = cv2.imread('Self_Driving_Car.jpg')
cv2.imshow('Original Image', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
height, width = image.shape[:2] # Get the heigh and width of colored image (3 layers, just want first 2)
```

# PERFORM ROTATIONS

```python
M_rotation = cv2.getRotationMatrix2D((width/2, height/2), 90, 0.5) # Rotate around center hence x/2, 90 degrees, resized by 0.5
rotated_image = cv2.warpAffine(image, M_rotation, (width, height)) # insert size of the output image

cv2.imshow('Rotated', rotated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

# PERFORM TRANSLATIONS

```python
# Translate 300 pixels x direction (right) and 50 pixels y direction (down)
Translational_Matrix = np.float32([[1, 0, 300], 
                                  [0, 1, 50]])
```

```python
translated_image = cv2.warpAffine(image, Translational_Matrix, (width, height))
cv2.imshow('Translated Image', translated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Don't worry about the color, recall openCV uses BGR instead of RGB
plt.imshow(translated_image)
```

# PERFORM RESIZING

```python
resized_image = cv2.resize(image, None, fx=2, fy=2,interpolation = cv2.INTER_CUBIC) # Resize image with fx, fy being the multiplier
cv2.imshow('Resized Image', resized_image) 
cv2.waitKey()
cv2.destroyAllWindows()
```
