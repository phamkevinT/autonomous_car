# Let's convert from RGB to HSV (Hue, Saturation, Value)

```python
image = cv2.imread('Self_Driving_Car.jpg')
cv2.imshow('Self Driving Car!', image) # Display colored image

hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2.imshow('HSV Image', hsv_image) # Display HSV image

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(hsv_image[:, :, 0]) # Display the first layer (Hue)
plt.title('Hue channel')
```

```python
plt.imshow(hsv_image[:, :, 1]) # Display the second layer (Saturation)
plt.title('Saturation channel')
```

```python
plt.imshow(hsv_image[:, :, 2]) # Display the third layer (Value)
# add a title to the image
plt.title('Value channel')
```

# Let's split and merge R, G, B channels 

```python
image = cv2.imread('Self_Driving_Car.jpg')
B, G, R = cv2.split(image) # Split the layers
B.shape
```

```python
R.shape
```

```python
# Let's view the R, B, G channels
cv2.imshow("Blue Channel!", B) 
cv2.waitKey(0)
cv2.destroyAllWindows()

# We didn't expect these Results!
# We get a grayscale image because opencv treat it as a single layer 0 - 255 value image
```

```python
# Let's try to create our own 3D image out of the blue channel!
zeros = np.zeros(image.shape[:2], dtype = "uint8") # Set the other channels to zero

cv2.imshow("Blue Channel!", cv2.merge([B, zeros, zeros]))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Let's merge RGB channels to create our original image
image_merged = cv2.merge([B, G, R]) 
cv2.imshow("Merged Image!", image_merged) 

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Let's merge our original image while adding more green! 
image_merged = cv2.merge([B, G+100, R]) # Added extra value to the green channel
cv2.imshow("Merged Image with some added green!", image_merged) 

cv2.waitKey(0)
cv2.destroyAllWindows()
```
